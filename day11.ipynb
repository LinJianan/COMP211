```python
# overload

class Time:
    def __init__(self, hour = 0, minute = 0, second = 0):
        self.hour = hour
        self.minute = minute
        self.second = second
    
    # this is a function supported by python
    # we can re-write this function to print what we want
    def __str__(self):
        # return str(self.hour) + ":" + str(self.minute) + ":" + str(self.second)
        # return 'hello world'
        result = ""
        if self.hour < 10:
            result = result + str(0) + str(self.hour)
        else:
            result = result + str(self.hour)
        result = result + ":"
        if self.minute < 10:
            result = result + str(0) + str(self.minute)
        else:
            result = result + str(self.minute)
        result = result + ":"
        if self.second < 10:
            result = result + str(0) + str(self.second)
        else:
            result = result + str(self.second)
        return result

    # compare equal
    def __eq__(self, other) -> bool:
        return self.hour == other.hour and self.minute == other.minute and self.second == other.second

    # compare not equal
    def __ne__(self, other) -> bool:
        print('we are comparing whether they are not equal')
        return not self.__eq__(other)

    # less than
    def __lt__(self, other) -> bool:
        if self.hour < other.hour:
            return True
        elif self.hour > other.hour:
            return False
        else:
            if self.minute < other.minute:
                return True
            elif self.minute > other.minute:
                return False
            else:
                return self.second < other.second

    # less than or equal to
    def __le__(self, other) -> bool:
        if self.hour < other.hour:
            return True
        elif self.hour > other.hour:
            return False
        else:
            if self.minute < other.minute:
                return True
            elif self.minute > other.minute:
                return False
            else:
                # this line is the only difference between le and lt
                return self.second <= other.second

    # greater than
    def __gt__(self, other) -> bool:
        return not self.__le__(other)

    # greater than or equal to
    def __ge__(self, other) -> bool:
        return not self.__lt__(other)

    # add +
    def __add__(self, other) -> bool:
        # we allow other be an integer representing second
        # also we allow other be another class time
        result = Time(self.hour, self.minute, self.second)
        flag = 0
        if type(other) == int:
            result.second = result.second + other
            # notice second < 60, flag is the minute we get
            flag = flag + result.second // 60
            result.second = result.second % 60
            result.minute = result.minute + flag
            # what if minute >= 60
            flag = 0
            flag = flag + result.minute // 60
            result.minute = result.minute % 60
            result.hour = result.hour + flag
            result.hour = result.hour % 24
            return result
        # other is a time class
        elif type(other) == Time:
            flag = 0
            result = Time(self.hour + other.hour, self.minute + other.minute, self.second + other.second)
            flag = result.second // 60
            result.second = result.second % 60
            result.minute = result.minute + flag
            flag = result.minute // 60
            result.hour = result.hour + flag
            result.minute = result.minute % 60
            return result
        else:
            print('We can not add these two together')
            return self
    
    # minus -
    def __sub__(self, other):
        if type(other) == int:
            return self + (-other)
        elif type(other) == Time:
            return self + Time(-other.hour, -other.minute, -other.second)
        else:
            print('We cannot do this sub operation')
            return self

    # also you should know the meaning of the following operator overload
    # __mul__        ->    *
    # __truediv__    ->    /
    # __floordiv__   ->    //
    # __mod__        ->    %
    # __pow__        ->    **
    # __and/or__     ->    and/or


time1 = Time(1, 2, 3)
print(time1)

time2 = Time(1, 2, 3)
print(time1 == time2)

time3 = Time(1, 2, 4)
print(time1 == time3)

print(time1 != time2)

print(time1 < time2)
print(time1 < time3)

print(time1 <= time2)
print(time1 <= time3)

print(time1 + 1)
print(time1 + 61)

print(time1 + time2)

print(time1 - time2)
```

```python
class Vector:

    # we assume the input is always valid
    # row vector = [1, 2, 3, 4, 5]
    # column vector = [[1], [2], [3], [4], [5]]
    def __init__(self, input: list):
        self.vector = input.copy()
        self.type = ''
        self.size = len(input)
        # we need to determine type
        # this is row vector
        if type(input[0]) == int:
            self.type = 'R'
        else:
            self.type = 'C'

    # then we implement the basic operations
    def print(self):
        if self.type == 'R':
            print(self.vector)
        else:
            print('[' + str(self.vector[0]) + ',')
            for i in range(1, self.size - 1):
                print(str(self.vector[i]) + ',')
            print(str(self.vector[self.size - 1]) + ']')

    def __str__(self):
        if self.type == 'R':
            return str(self.vector)
        else:
            result = ""
            result = result + '[' + str(self.vector[0]) + ',\n'
            for i in range(1, self.size - 1):
                result = result + str(self.vector[i]) + ',\n'
            result = result + str(self.vector[self.size - 1]) + ']'
            return result
    
    # still we only consider the input is valid
    def plus(self, other):
        if self.type == 'R':
            result = [0] * self.size
            for i in range(self.size):
                result[i] = self.vector[i] + other.vector[i]
            return Vector(result)
        else:
            result = [[0] for i in range(self.size)]
            for i in range(self.size):
                result[i][0] = self.vector[i][0] + other.vector[i][0]
            return Vector(result)

    def __add__(self, other):
        if self.type == 'R':
            result = [0] * self.size
            for i in range(self.size):
                result[i] = self.vector[i] + other.vector[i]
            return Vector(result)
        else:
            result = [[0] for i in range(self.size)]
            for i in range(self.size):
                result[i][0] = self.vector[i][0] + other.vector[i][0]
            return Vector(result)

    def __sub__(self, other):
        sub_list = [-e for e in other.vector]
        return self + Vector(sub_list)


vector1 = Vector([1, 2, 3, 4, 5])
vector2 = Vector([[1], [2], [3], [4], [5]])
print(vector1)
print(vector2)

vector3 = Vector([6, 7, 8, 9, 10])
print(vector1 + vector3)
print(vector3 - vector1)
```

```python
import copy

class Matrix:
    def __init__(self, input):
        self.matrix = input.copy()
        self.M = len(input)
        self.N = len(input[0])

    def print(self):
        if self.M == 1:
            print(self.matrix)
            return
        print('[' + str(self.matrix[0]) + ',')
        for i in range(1, self.M - 1):
            print(str(self.matrix[i]) + ',')
        print(str(self.matrix[self.M - 1]) + ']')

    def __str__(self):
        if self.M == 1:
            return str(self.matrix)
        else:
            result = ""
            result = result + '[' + str(self.matrix[0]) + ',\n'
            for i in range(1, self.M - 1):
                result = result + str(self.matrix[i]) + ',\n'
            result = result + str(self.matrix[self.M - 1]) + ']'
            return result
    
    def plus(self, other):
        result = self.matrix.copy()
        for i in range(self.M):
            for j in range(self.N):
                result[i][j] = self.matrix[i][j] + other.matrix[i][j]
        return Matrix(result)

    def __add__(self, other):
        result = copy.deepcopy(self.matrix)
        for i in range(self.M):
            for j in range(self.N):
                result[i][j] = self.matrix[i][j] + other.matrix[i][j]
        return Matrix(result)

    def __sub__(self, other):
        result = copy.deepcopy(self.matrix)
        for i in range(self.M):
            for j in range(self.N):
                result[i][j] = self.matrix[i][j] - other.matrix[i][j]
        return Matrix(result)

    def multiply(self, other):
        M, N, L = self.M, other.N, self.N # other.M is also OK
        result = [[0 for j in range(N)] for i in range(M)]
        for i in range(M):
            for j in range(N):
                for k in range(L):
                    result[i][j] = result[i][j] + self.matrix[i][k] * other.matrix[k][j]
        return Matrix(result)

    def __mul__(self, other):
        M, N, L = self.M, other.N, self.N # other.M is also OK
        result = [[0 for j in range(N)] for i in range(M)]
        for i in range(M):
            for j in range(N):
                for k in range(L):
                    result[i][j] = result[i][j] + self.matrix[i][k] * other.matrix[k][j]
        return Matrix(result)

    def transpose(self):
        M, N = self.N, self.M
        result = [[0 for j in range(N)] for i in range(M)]
        for i in range(M):
            for j in range(N):
                result[i][j] = self.matrix[j][i]
        return Matrix(result)

    # determinant requires an N * N matrix, and suppose input is always valid
    def determinant(self):
        matrix = self.matrix
        # slice is to cut the matrix and get a smaller one
        def slice(matrix, start_row, end_row, start_col, end_col):
            result = []
            for i in range(start_row, end_row + 1):
                result.append([])
                for j in range(start_col, end_col + 1):
                    result[i - start_row].append(matrix[i][j])
            return result
        # we use recursion to implement determinant calculation
        def determinat_helper(matrix):
            N = len(matrix)
            if N == 1:
                return matrix[0][0]
            elif N == 2:
                return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
            else:
                flag = 1
                result = 0
                # this is the same as formula of determinant
                for i in range(N):
                    result = result + flag * matrix[i][0] * determinat_helper(slice(matrix, 0, i - 1, 1, N - 1) \
                        + slice(matrix, i + 1, N - 1, 1, N - 1))
                    # rememebr next time the operator is opposite, + - + - + - ...
                    flag = flag * (-1)
                return result
        return determinat_helper(matrix)

matrix1 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(matrix1)

matrix2 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(matrix1 + matrix2)
print(matrix1 - matrix2)

print(matrix1 * matrix2)
```
